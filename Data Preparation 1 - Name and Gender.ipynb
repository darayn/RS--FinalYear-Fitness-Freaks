```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
def parse_url(url):
    page  = requests.get(url).text
    return bs(page)
```

```python
girl = parse_url('https://www.in.pampers.com/pregnancy/baby-names/article/top-indian-baby-names')
```

```python
girls_list_table = girl.find_all('table')[0]
```

```python
girls_list_table
```

```python
tds_girls=girls_list_table.find_all('tr')
tds_girls
```

```python
sample=tds_girls[1].find('td').text
```

```python
sample
```

```python
girls_list = []
for i in range(1,50):
    girls_list.append(tds_girls[i].find('td').text)
```

```python
girls_list
```

```python
boys_list_table = girl.find_all('table')[1]
```

```python
tds_boys=boys_list_table.find_all('tr')
tds_boys
```

```python
boys_list = []
for i in range(1,52):
    boys_list.append(tds_boys[i].find('td').text)
```

```python
boys_list
```

```python
surname_page = parse_url('https://www.schoolmykids.com/parenting/popular-indian-last-names-indian-surnames')
```

```python
surrnames = surname_page.find(text = re.compile('10 Best Indian Family Movies of All Time')).next.next.next.next
```

```python


surrnames.find_all('li')[0]

```

```python
surname_list=[]
for i in range(0,102):
    surname_list.append(surrnames.find_all('li')[i].text)
```

```python
surname_list
```

```python
final_surnames = []
for i in surname_list:
    value = i.split()[0]
    final_surnames.append(value)
```

```python
final_surnames
```

```python
first_names=boys_list+girls_list
# last_names=('Johnson','Smith','Williams')
```

```python
len(boys_list), len(girls_list)
```

```python
len(first_names)
```

```python
def convert(list):
    return tuple(list)
```

```python
first_names = convert(first_names)
```

```python
surnames = convert(final_surnames)
```

```python
name_gender= {"".join(random.choice(first_names)+" "+random.choice(surnames)):('Male') for _ in range(1500)}
```

```python
len(name_gender)
```

```python
for key in name_gender.keys():
    if key.split()[0] in girls_list:
        name_gender[key] = 'Female'
```

```python
name_gender
```

## Creating Data Frame from dictionary

```python
s = pd.Series(name_gender, name='Gender')
```
