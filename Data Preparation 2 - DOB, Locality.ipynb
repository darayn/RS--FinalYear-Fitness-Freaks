```python
import numpy as np
import pandas as pd
import re
import random
from bs4 import BeautifulSoup as bs
import requests
import datetime
```

```python
start_date = datetime.date(1981, 1, 1)
end_date = datetime.date(2003, 12, 31)

def generate_date(start_date,end_date):
    time_between_dates = end_date - start_date
    days_between_dates = time_between_dates.days
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start_date + datetime.timedelta(days=random_number_of_days)
    return random_date
```

## age considered here is 18 - 40

```python
date = generate_date(start_date,end_date)
```

```python
date
```

```python
Dob_list = []
for i in range(1387):
    date_ = generate_date(start_date,end_date)
    Dob_list.append(date_)
```

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
def parse_url(url):
    page  = requests.get(url).text
    return bs(page)
```
