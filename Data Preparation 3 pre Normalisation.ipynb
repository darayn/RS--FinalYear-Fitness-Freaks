```python
from datetime import date
from datetime import datetime
def calculateAge(birthDate):
    birthDateyear = int(birthDate.split('-')[0])
    birthDatemonth = int(birthDate.split('-')[1])
    birthDateday = int(birthDate.split('-')[2])
    today = date.today()
    age = today.year - birthDateyear - ((today.month, today.day) < (birthDatemonth, birthDateday))  
    return age
```

```python
date_ex = '1982-03-31'
# datetime.strptime(date_ex, '%m/%d/%y')
int(date_ex.split('-')[0])
calculateAge(date_ex)
```
