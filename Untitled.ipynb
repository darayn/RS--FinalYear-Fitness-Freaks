## Dividing Datasets for Normalisation - Content Filtering

```python
df1 = df[['Name', 'Gender', 'Flexibility', 'Body Composition', 'Composition Percentage', 'Interactiveness', 'Workout Frequency', 'Workout Mode Preferred', 'is_sports_enthusiast', 'like_to_be_mentored', 'like_to_mentor', 'fitness_goal_to_be achieved','addons in session','Height_in_cms','Age']]
```

```python
df1
```

```python
df1['Gender'] = df1["Gender"].astype('category').cat.codes
```

```python
df1['Body Composition'] = df1["Body Composition"].astype('category').cat.codes
df1['Workout Frequency'] = df1["Workout Frequency"].astype('category').cat.codes
df1['Workout Mode Preferred'] = df1["Workout Mode Preferred"].astype('category').cat.codes
df1['is_sports_enthusiast'] = df1["is_sports_enthusiast"].astype('category').cat.codes
df1['fitness_goal_to_be achieved'] = df1["fitness_goal_to_be achieved"].astype('category').cat.codes
df1['addons in session'] = df1["addons in session"].astype('category').cat.codes
df1
```

```python
df1
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df1_scaled = df1.copy() 
names = ['Flexibility','Composition Percentage','Interactiveness', 'like_to_be_mentored', 'like_to_mentor', 'Height_in_cms', 'Age']

df1_scaled[names] = MinMaxScaler().fit_transform(df1_scaled[names])
```

```python
df1_scaled
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(df1_scaled.drop('Name',axis=1))
print(similarity)
```

```python
df1_similarity = pd.DataFrame(similarity, columns=df1_scaled['Name'], index=df1_scaled['Name'])
```
